```python
import os
import json
import pandas as pd
from dotenv import load_dotenv
from serpapi import GoogleSearch
```

```python
# Get env variables from .env file
load_dotenv()

# Get the API KEY 
api_key = os.getenv('SERPAPI_API_KEY')

# Configure the search params
params = {
    "engine": "google_jobs",
    "q": "Data Engineer",
    "location": "United States",
    "api_key": api_key
}

# Create a GoogleSearch instance and make the search
search = GoogleSearch(params)
results = search.get_dict()
display(results)
```

```python
normalize_results = pd.json_normalize(results)

# We check all the response
df = pd.DataFrame(normalize_results)
df.rename(columns={'serpapi_pagination.next_page_token':'next_page_token'}, inplace=True)
display(df.columns)
```

```python
total_jobs = 0

# We get all the jobs
while total_jobs < 1:    
    jobs = results.get('jobs_results', [])
    
    for job_result in jobs:
        total_jobs += 1

    next_page_token = df['next_page_token']
    
    if next_page_token.empty:
        break

    params['next_page_token'] = next_page_token
    

print(f"Total jobs so far: {total_jobs}")
```
